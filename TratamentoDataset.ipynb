#### Data Cleaning

```python
# %pip install nltk
```

```python
import re
from tqdm import tqdm
import spacy
import en_core_web_sm
import pt_core_news_sm
import nltk
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))
stopword_pt = nltk.corpus.stopwords.words('portuguese')
```

```python
def clean_en(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub(r'\b\w{1,2}\b', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    return text

def clean_pt(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub(r'\b\w{1,2}\b', '', text)
    text = [word for word in text.split(' ') if word not in stopword_pt]
    text=" ".join(text)
    return text

df["tweet"] = df["tweet"].apply(clean_en)
df["tweets_pt"] = df["tweets_pt"].apply(clean_en)
```

```python
def lemmatize_text(text):
    doc = pt(text)
    lemmas = [token.lemma_ for token in doc]
    return " ".join(lemmas)
# df['lemmatized_text'] = df['tweet'].apply(lemmatize_text)

for i, row in tqdm(df.iterrows(), total=len(df)):
    lemmatized_text = lemmatize_text(row['tweets_pt'])
    df.at[i, 'pt_lemma'] = lemmatized_text
```
