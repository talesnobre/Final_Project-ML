```python
from nltk.util import pr
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
# Importando os dados
df = pd.read_csv('./Dados/twitter.csv')
df.head()
```

```python
x = np.array(df["tweets_pt"])
y = np.array(df["label"])
```

```python
cv = CountVectorizer()
X = cv.fit_transform(x) # Fit the Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.metrics import classification_report, accuracy_score

clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
clf.score(X_test,y_test)

print('Ein = %0.4f' % (1 - clf.score(X_train,y_train)))
print('Eout = %0.4f' % (1 - clf.score(X_test,y_test)))
print(classification_report(y_test, clf.predict(X_test)))
```

```python

for feature, importance in zip(cv.get_feature_names_out(), clf.feature_importances_):
    print("{}:{}".format(feature, importance))

```
