```python
df = pd.read_csv("twitter_st.csv")
df
```

```python
from tqdm import tqdm
from mtranslate import translate

translated_tweets = []

for tweet in tqdm(df['tweet']):
    translated_tweet = translate(tweet, 'pt', 'en')
    translated_tweets.append(translated_tweet)

df['tweets_pt'] = translated_tweets
df.to_csv('Dados/twitter_trad.csv', index=False)
```
